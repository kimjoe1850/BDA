# 필수과제 1
- sklearn 공식홈페이지에 파라미터들에 대한 자세한 설명이 나와있음.
- 파라미터들 개념 정리하기

## penalty : {‘l1’, ‘l2’, ‘elasticnet’, None}, default=’l2’
- penalty 설정, ‘l1’, ‘l2’, ‘elasticnet’, None의 4가지가 존재함
  - None : 어떠한 패털티도 설정되지 않음

  - 'l2': L2 penalty 설정, 디폴트가 l2

  - 'l1': L1 penalty 설정

  - 'elasticnet': L1, L2 penalty 두개 다 적용되는 엘라스틱넷 설정

- 특정 penalty는 몇몇 solver와 작동하지 않을 수 있다

- 0.19버전 업데이트 :  l1 penalty + SAGA solver를 통해 ‘multinomial’ + L1 가능
- 1.2버전에서 ‘none'은 사용되지 않으며 1.4버전에서 제거됨. None을 대신 사용

## l1_ratio : float(실수), default=None
- penalty로 elasticnet을 사용할 경우에만 사용하는 파라미터로, 0~1 사이의 실수값
- 부과할 l1 penalty의 비율을 나타낸다(elasticnet = L1 + L2)
- l1_ratio = 0이면 penalty = 'l2'와 같고, 1인 경우 penalty = 'l1'과 같다

## C : float(실수), default=1.0
- 규제의 강도를 조절하는 파라미터로, 양의 실수값을 가진다  
- SVM처럼 작은 값일수록 더 강한 규제를 가진다

## class_weight : dict(딕셔너리) or ‘balanced’, default=None
- 학습 시 class에 따라 가중치를 다르게 주고 싶을 때 사용할 수 있는 파라미터
- 딕셔너리 형태로 {class_label: weight} 직접 적용 가능하거나, 적용하지 않을 경우 모든 class의 가중치가 1이 된다
- “balanced”를 적용할 경우 (전체 샘플 수) / (클래스 수 * 클래스별 빈도)로 계산.

## solver : {‘lbfgs’, ‘liblinear’, ‘newton-cg’, ‘newton-cholesky’, ‘sag’, ‘saga’}, default=’lbfgs’
- 최적화 문제를 풀기 위한 알고리즘을 선택하는 파라미터로, 디폴트값은 ‘lbfgs’이다.
- To choose a solver, you might want to consider the following aspects:
- 데이터의 양이 작은 경우 ‘liblinear’, 데이터 양이 많은 경우는 ‘sag’와 ‘saga’가 더 빠르다
- 다중분류 문제의 경우, ‘newton-cg’, ‘sag’, ‘saga’, ‘lbfgs’ 만이 multinomial loss를 다룰 수 있다
- ‘liblinear’은 one-versus-rest방식으로 제한된다.
- ‘newton-cholesky’는  (샘플의 수) >> (feature의 수)인 경우에 사용하기 좋은데, 특히 rare category의 원핫 인코딩된 범주형 features에 좋다.

## random_state : int(정수), RandomState instance, default=None
- solver == ‘sag’, ‘saga’, ‘liblinear’일 때 cycle등의 문제가 발생하지 않도록 데이터를 shuffle하게 되는데, 이때의 randomness를 고정

## max_iter : int(정수), default=100
- solver에 의해 진행되는 수렴을 위한 반복의 최대 횟수 지정

## multi_class : {‘auto’, ‘ovr’, ‘multinomial’}, default=’auto’
- ‘ovr’ : one vs rest 방식으로 각 클래스 레이블에 대해 이진 분류 문제를 푼다
- ‘multinomial’ : 각 클래스에 대한 softmax 확률값, 크로스 엔트로피(cross-entropy) 계산을 통해 원핫 타겟 벡터를 생성하는 방법으로 다중분류

## n_jobs : int(정수형), default=None
- multi_class=’ovr’일때 모델 작동할 때 동시에 사용할 CPU코어 수
- solver가 ‘liblinear’인 경우 ‘multi_class’가 정해져있어도 n_jobs 파라미터는 무시한다
- 1인 경우 None, -1인 경우 모든 프로세서를 사용한다

## fit_intercept : bool, default=True
- 모형에 intercept(상수항, 절편)이 있는지 결정

##dual : bool, default=False
- 이중 또는 초기 공식
- 이중 공식은 l2 penalty와 'liblinear' solver일 때만 구현됨.
- n_samples > n_features인 경우 dual=False를 선호

## intercept_scaling : float, default=1
- 정규화의 효과 정도를 설정한다
- solver가 ‘liblinear’이고 self.fit_intercept가 True로 설정된 경우 용이하다.

## tol : float, default=1e-4
- 정밀도

## verbose : int, default=0
- liblinear and lbfgs solver에서 동작 과정에 대한 출력 메세지

## warm_start : bool, default=False
- 이전 모델을 초기화로 적합하게 사용할 것인지의 여부
