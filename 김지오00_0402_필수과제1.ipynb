```python
# 데이터 불러오기
import pandas as pd
df = pd.read_csv('./wine_test.csv')
df_sp = df[['alcohol', 'ash', 'nonflavanoid_phenols' ,'proanthocyanins']]
df_sp.dropna(inplace = True)
```

# 필수과제 1

- from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler

- 스케일링 진행하지 않은 2개 + 추가로 더 하셔도 됩니다 abs, 등으로 차이를 비교해 주세요!

## MinMaxScaler()
- Min-Max Normalization 이라고도 불리며, 특성들을 특정 범위(주로 [0,1]) 로 스케일링 하는 것
- 가작 작은 값은 0, 가장 큰 값은 1 로 변환되므로, 모든 특성들은 [0, 1] 범위를 가짐.
- 이상치에 매우 민감함
- 분류보다 회귀에 유용함

```python
# 모델 생성
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, random_state=111)
```

```python
from sklearn.preprocessing import MinMaxScaler

minmax_scaler = MinMaxScaler()
minmax_scaler.fit(df_sp)
df_sp_minmax = minmax_scaler.transform(df_sp)
```

```python
kmeans.fit(df_sp_minmax)
```

```python
# 스케일링 결과 데이터프레임으로 만들기
df_sp_mm = pd.DataFrame(df_sp_minmax, columns = df_sp.columns)
```

```python
df_sp_mm.head()
```

```python
print(kmeans.labels_)
```

```python
df_sp_mm['cluster'] = kmeans.labels_
df_sp_mm['class'] = df['class']
```

```python
# 기존 클래스별 분포
df.groupby('class')['alcohol'].count()
```

```python
# 군집화 한 이후의 분포
df_sp_mm.groupby('cluster')['alcohol'].count()
```

## MaxAbsScaler()
- 각 특성의 절대값이 0 과 1 사이가 되도록 스케일링
- 즉, 모든 값은 -1 과 1 사이로 표현, 데이터가 양수일 경우 MinMaxScaler 와 같다
- 이상치에 매우 민감함


```python
from sklearn.preprocessing import MaxAbsScaler

maxabs_scaler = MaxAbsScaler()
maxabs_scaler.fit(df_sp)
df_sp_maxabs = maxabs_scaler.transform(df_sp)
```

```python
# 모델 생성
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, random_state=111)
kmeans.fit(df_sp_maxabs)
```

```python
# 스케일링 결과 데이터프레임으로 만들기
df_sp_ma = pd.DataFrame(df_sp_maxabs, columns = df_sp.columns)
```

```python
df_sp_ma.head()
```

```python
print(kmeans.labels_)
```

```python
df_sp_ma['cluster'] = kmeans.labels_
df_sp_ma['class'] = df['class']
```

```python
# 기존 클래스별 분포
df.groupby('class')['alcohol'].count()
```

```python
# 군집화 한 이후의 분포
df_sp_ma.groupby('cluster')['alcohol'].count()
```

## RobustScaler()
- 평균과 분산 대신 중간값과 사분위 값을 사용
  - 중간값 : 정렬시 중간에 있는 값
  - 사분위값 : 1/4, 3/4에 위치한 값
- 이상치 영향을 최소화할 수 있음

```python
from sklearn.preprocessing import RobustScaler

robust_scaler = RobustScaler()
robust_scaler.fit(df_sp)
df_sp_robust = robust_scaler.transform(df_sp)
```

```python
# 모델 생성
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, random_state=111)
kmeans.fit(df_sp_robust)
```

```python
# 스케일링 결과 데이터프레임으로 만들기
df_sp_rb = pd.DataFrame(df_sp_robust, columns = df_sp.columns)
```

```python
df_sp_rb.head()
```

```python
print(kmeans.labels_)
```

```python
df_sp_rb['cluster'] = kmeans.labels_
df_sp_rb['class'] = df['class']
```

```python
# 기존 클래스별 분포
df.groupby('class')['alcohol'].count()
```

```python
# 군집화 한 이후의 분포
df_sp_rb.groupby('cluster')['alcohol'].count()
```

## Normalizer()
- 앞에서 나온 스케일러는 각 특성(열)의 통계치를 이용하여 진행
- Normalizer는 각 샘플(행)마다 적용
  - 한 행의 모든 특성들 사이의 유클리드 거리(L2 norm)가 1이 되도록 스케일링
- 일반적인 데이터 전처리의 상황이 아닌 모델(특히 딥러닝) 내 학습 벡터에 적용
  - 특히 피쳐들이 다른 단위(키, 나이, 소득 등)라면 더욱 사용하지 않음

```python
from sklearn.preprocessing import Normalizer

normal_scaler = Normalizer()
normal_scaler.fit(df_sp)
df_sp_normal = normal_scaler.transform(df_sp)
```

```python
# 모델 생성
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, random_state=111)
kmeans.fit(df_sp_normal)
```

```python
# 스케일링 결과 데이터프레임으로 만들기
df_sp_nm = pd.DataFrame(df_sp_normal, columns = df_sp.columns)
```

```python
df_sp_nm.head()
```

```python
print(kmeans.labels_)
```

```python
df_sp_nm['cluster'] = kmeans.labels_
df_sp_nm['class'] = df['class']
```

```python
# 기존 클래스별 분포
df.groupby('class')['alcohol'].count()
```

```python
# 군집화 한 이후의 분포
df_sp_nm.groupby('cluster')['alcohol'].count()
```

# 전체 비교

```python
scaler = [df_sp_mm, df_sp_ma, df_sp_rb, df_sp_nm]
for scale in scaler:
  print("MinMaxScaler, MaxAbsScaler, RobustScaler, Normalizer의 군집화 결과")
  print("############################")
  print(scale.groupby('cluster')['alcohol'].count())
```

# 추가 : 언제 스케일링을 해야 하는가
- 스케일링을 해야 하는 경우
  - 주성분 분석(PCA) : 높은 분산/넓은 범위 가지는 변수는 주성분에서 큰 회귀계수를 가짐. 이는 중요하지 않은 변수도 범위가 넓으면 주성분에서 중요한 변수로 간주되므로 모델링 수행 시 치명적
  - 군집(clustering) : 거리 기반 알고리즘으로 거리 측도를 이용해 관측값 사이에서 유사성을 찾아 군집 형성. 따라서 넓은 범위를 가지는 변수는 군집에서 더 큰 영향력을 가짐
  - knn : 거리 기반 알고리즘으로 새로운 관측값의 주변에 있는 k개의 이웃을 이용해 유사성 측도에 기반해 관측값 분류. 따라서 유사성 측도에 모든 변수가 동일하게 기여하도록 스케일링 수행해야 함
  - svm : 서포트벡터와 분류기 사이 거리(마진)을 최대로 만들어주는 분류기 찾는 알고리즘. 즉 큰 값을 가지는 변수가 거리 계산 시 영향 많음.
  - 회귀 : 예측을 위한 회귀 모델링이라면 표준화 수행할 필요 없음. 변수의 중요도나 다른 회귀계수들과 비교를 하는것이 목적이라면 표준화 수행해야 함.
  - ridge / lasso : 회귀계수에 벌점을 가하는 방법으로 다중공선성, 과적합 해결하기 위해 사용. 회귀계수의 크기를 벌점으로 사용하므로 상대적으로 큰 분산을 가지는 변수가 회귀계수가 클 것이고 작은 분산을 가지는 변수들보다 회귀계수를 더 축소시켜 최종적 회귀계수는 작아지게 됨. 즉 회귀계수가 단위에 영향을 미치므로 표준화 수행해야 함

- 스케일링이 필요하지 않은 경우
  - 로지스틱 회귀, 트리 기반 의사결정나무, 랜덤포레스트, 그래디언트 부스팅
  - 변수의 크기에 민감하지 않아 표준화 수행할 필요 없음
