3차 다항식을 만들고 원리를 주석으로 간단하게 정리하기

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
# 다항회귀가 변하는 방식
X = np.arange(6).reshape(2,3)
print('1차 단항삭의 계수 피처 ','\n',X)
```

```python
# 2차, 3차로 바뀌는 방식 이해
poly = PolynomialFeatures(degree=3) # 3차 다항식
poly.fit(X)
poly_ftr = poly.transform(X)
print('변환된 3차 다항식 계수','\n',poly_ftr)
```

- 단항식은 [X1, X2, X3]
- 3차 다항식으로 바꾸는 것은 기존의 단항식을 가지고 다항계수 변경
- [1, X1, X2, X3, X1^2, X1X2, X1X3, X2^2, X2X3, X3^2, X1^3, X1^2X2, X1^2X3, X2^2X1, X1X2X3, X3^2X1, X2^3, X2^X3, X3^2X2, X3^3]
