# 엔트로피(Entropy)
image.png

- 열역학에서 나온 개념으로 무질서에 대한 측도 역할, 확률 변수의 불확실성을 수치로 나타낸 것

  - 엔트로피가 1에 가까울수록 출력값은 동등한 확률(아무런 정보를 갖고 있지 않음)

# 정보 이익(IG : Information Gain)
image.png
- 정보의 가치를 의미하며 값이 클수록 좋음
- 정보이익 = 사전 엔트로피(불확실성) - 사후 엔트로피(불확실성)
- IG는 불확실성이 얼마나 줄어들었는가?로 해석할 수 있으며, 값이 클수록 불확실성이 많이 줄어들었다고 해석 가능

# 이득율(IGR : Information Gain Ratio)

## IV(Intrinsic Value)
image.png

## 이득율(IGR)
image.png
- 이득률(IGR)은 정보이익(IG)을 IV로 나눈 값
- IV에 개수를 넣어 그 개수에 따라 패널티 부여
- 패널티에 따라 값이 조정되어 IG의 단점 보완

```python
# entropy 식
def entropy(x):
    return (-x*np.log2(x)).sum()
```

```python
# 샘플 데이터 생성
group_1 = np.array([0.2,0.4,0.3])
group_2 = np.array([0.2,0.9,0.4])
group_3 = np.array([0.02,0.97,0.01])
```

```python
fig, axes = plt.subplots(1, 3)
fig.set_size_inches(12, 4)
axes[0].bar(np.arange(3), group_1, color='blue')
axes[0].set_title('Group 1')
axes[1].bar(np.arange(3), group_2, color='red')
axes[1].set_title('Group 2')
axes[2].bar(np.arange(3), group_3, color='green')
axes[2].set_title('Group 3')
plt.show()
```

```python
# 엔트로피 계산 및 시각화
entropy_1 = entropy(group_1)
entropy_2 = entropy(group_2)
entropy_3 = entropy(group_3)

print(f'Group 1: {entropy_1:.3f}\nGroup 2: {entropy_2:.3f}\nGroup 3: {entropy_3:.3f}')
```

```python
plt.figure(figsize=(5, 5))
plt.bar(['Group 1', 'Group 2', 'Group 3'], [entropy_1, entropy_2, entropy_3])
plt.title('Entropy', fontsize=15)
plt.show()
```

- Group1의 엔트로피가 가장 크고, Group3의 엔트로피가 가장 작다
- 따라서 Group3의 출력이 가장 확실하고 질서있다고 할 수 있고, Group1의 출력이 가장 불확실하고 무질서하다고 할 수 있다

# 지니계수
image.png
- 불순도를 측정하는 지표
- 데이터의 통계적 분산 정도를 정량화하여 표현한 값
- 지니계수가 높을수록 데이터가 분산되었음을 의미
- 지니계수가 높을수록 잘 분류되지 못한 것이라고 할 수 있따

```python
# 지니계수
def gini(x):
    return 1 - ((x / x.sum())**2).sum()
```

```python
# 지니계수 구현 및 시각화
gini_1 = gini(group_1)
gini_2 = gini(group_2)
gini_3 = gini(group_3)

print(f'Group 1: {gini_1:.3f}\nGroup 2: {gini_2:.3f}\nGroup 3: {gini_3:.3f}')
```

```python
plt.figure(figsize=(5, 5))
plt.bar(['Group 1', 'Group 2', 'Group 3'], [gini_1, gini_2, gini_3])
plt.title('Gini Index', fontsize=15)
plt.show()
```

- 엔트로피와 마찬가지로 Group1의 지니계수가 가장 크고, Group3의 지니계수가 가장 작다
- 따라서 Group3의 데이터의 분산이 가장 작고, Group1의 데이터의 분산이 크다는 것을 알 수 있다
